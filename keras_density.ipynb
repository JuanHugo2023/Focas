Examine which layers are frozen, and which are trainable

```python
for (i,layer) in enumerate(model.layers):
    print(i,layer.name,layer.trainable)
```

I've tried fine tuning lower layers after training to convergance on the top layers. I didn't see any dramatic improvements, but this is how you can unfreeze some lower layers

```python
for layer in model.layers:
    layer.trainable = layer.name.startswith("block14")
    layer.trainable |= layer.name.startswith("block15")
```

An example I saw switched so SGD with a much lower learning rate when fine tuning lower layers, like this

## Single image

```python
tid=0
pred=predict_large_image(model,engine.training_mmap_image(tid))
pred_sum = pred.sum(axis=(0,1))
true_sum = engine.counts.loc[tid,engine.class_names].as_matrix()
abs_diff = np.abs(pred_sum-true_sum)
print("pred:", pred_sum)
print("true:", true_sum)
print("diff:", abs_diff)
print("score:", np.mean(abs_diff))
```
